```python
import simplekml 

def parse_kml(file_path):
    kml = simplekml.Kml()
    kml.open(file_path)
    waypoints = []

    for placemark in kml.findall('Placemark'):
        coords = placemark.geometry.coords
        for coord in coords:
            waypoint = {
                'latitude': coord[1],
                'longitude': coord[0],
                'altitude': coord[2] if len(coord) > 2 else 0
            }
            waypoints.append(waypoint)

    return waypoints
```

```python
from pykml import parser

def parse_kml(file_path):
    with open(file_path, 'rb') as f:
        root = parser.parse(f).getroot()

    all_coordinates = []

    # Traverse through all Placemarks in the KML file
    for placemark in root.Document.iterfind('.//Placemark'):
        # Extract coordinates from each Placemark
        coordinates = str(placemark.Point.coordinates).strip().split(',')
        coordinates = list(map(float, coordinates))  # Convert coordinates to floats
        all_coordinates.append(coordinates)

    return all_coordinates
```

```python
from pykml import parser

class KMLParser:
    def __init__(self, file_path):
        self.file_path = file_path

    def parse(self):
        with open(self.file_path, 'rb') as f:
            root = parser.parse(f).getroot()

        waypoints = []

        for placemark in root.Document.Folder.Placemark:
            name = None
            if hasattr(placemark, 'name'):
                name = str(placemark.name)

            coords_str = str(placemark.Point.coordinates).strip()
            coords = [float(c) for c in coords_str.split(',')]

            waypoint = {
                'latitude': coords[1],
                'longitude': coords[0],
                'altitude': coords[2] if len(coords) > 2 else 0
            }
            waypoints.append(waypoint)

        return waypoints

# Example usage:

kml_parser = KMLParser(file_path_2)
waypoints = kml_parser.parse()

# Printing waypoints for verification
for wp in waypoints:
    print(wp)
```
